```python
import selenium
from selenium import webdriver
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from selenium.common.exceptions import NoSuchElementException
import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome("chromedriver.exe")
```

```python
url = "https://www.flipkart.com/"
driver.get(url)
```

```python
cross_bar = driver.find_element_by_xpath("//button[@class = '_2KpZ6l _2doB4z']")
cross_bar.click()
```

```python
search_bar = driver.find_element_by_xpath("//input[@class = '_3704LK']")
search_bar
```

```python
search_bar.send_keys("Laptops")
```

```python
search_loc=driver.find_element_by_xpath("//button[@class = 'L0Z3Pu']")
search_loc.click()
```

```python
reviews_of_the_product=[]
rating_of_the_product=[]
```

```python
url = []
for i in driver.find_elements_by_xpath("//a[@target = '_blank']"):
    url.append(i.get_attribute("href"))
url
```

```python
for j in range(300):
    for i in driver.find_elements_by_xpath("//div[@class= '_3LWZlK _1BLPMq']"):
        rating_of_the_product.append(i.text)
rating_of_the_product
```

```python
for j in range(300):
    for i in driver.find_elements_by_xpath("//div[@class= 't-ZTKy']"):
        reviews_of_the_product.append(i.text)
reviews_of_the_product
```

```python
print(len(rating_of_the_product))
print(len(reviews_of_the_product))
```
